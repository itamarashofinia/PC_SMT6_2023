# **Merubah warna menajadi Red, Green, Blue**

```python
import cv2 as cv 
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import drive

# Accessing my Google Drive
drive.mount('/content/drive')

# Load image
img = cv.imread('/content/drive/MyDrive/PC/smile.jpg')

# Splitting the channels
b, g, r = cv.split(img)

# Setting blue and green channel to 0
b = np.zeros_like(b)
g = np.zeros_like(g)

# Merge the channels
img_red = cv.merge([b, g, r])

# Display the image
cv2_imshow(img_red)
```

# **Membuat garis vertikal & horizontal**

```python
import cv2 as cv 
from google.colab.patches import cv2_imshow
from google.colab import drive

# Accessing my Google Drive
drive.mount('/content/drive')

# Load image
img = cv.imread('/content/drive/MyDrive/PC/smile.jpg')

# Draw a vertical line
cv.line(img, (img.shape[1]//2, 0), (img.shape[1]//2, img.shape[0]), (0, 0, 255), 5)

# Display the image with the line
cv2_imshow(img)
```

# **Lakukan flip vertikal & horizontal**

```python
import cv2 as cv 
from google.colab.patches import cv2_imshow
from google.colab import drive

# Accessing my Google Drive
drive.mount('/content/drive')

# Load image
img = cv.imread('/content/drive/MyDrive/PC/soal3.jpg')

# Flip vertically
img_flip_vert = cv.flip(img, 0)
cv2_imshow(img_flip_vert)

# Flip horizontally
img_flip_horiz = cv.flip(img, 1)
cv2_imshow(img_flip_horiz)
```
